```python
from tensorboard.backend.event_processing import event_accumulator
import numpy as np

def get_scalars(log_dir):

    # Event accumulator to load the logs
    ea = event_accumulator.EventAccumulator(log_dir,
        size_guidance={event_accumulator.SCALARS: 0})
    ea.Reload()

    scalars = {}
    for tag in ea.Tags()['scalars']:
        events = ea.Scalars(tag)
        steps = [e.step for e in events]
        values = [e.value for e in events]
        scalars[tag] = (steps, values)
    return scalars
```

```python
from tqdm import tqdm

dirs = ['/data/scpark/save/lse/train02.02-1',
        '/data/scpark/save/lse/train02.02-2',
        '/data/scpark/save/lse/train02.02-3',
        '/data/scpark/save/lse/train02.02-4']

lse_losses = []
recon_losses = []
for dir in tqdm(dirs):
    scalar = get_scalars(dir)
    lse_losses.append(scalar['lse_loss'])
    recon_losses.append(scalar['recon_loss'])

print('done')
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=[18, 4])
for i, lse_loss in enumerate(lse_losses[:1]):
    plt.plot(lse_loss[0][::100], lse_loss[1][::100], label=dirs[i])
plt.ylim([-2.6, -2.4])
plt.grid()
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=[18, 4])
for i, recon_loss in enumerate(recon_losses):
    plt.plot(recon_loss[0][::100], np.log10(recon_loss[1][::100]), label=dirs[i])
plt.grid()
plt.legend()
plt.show()
```
