```python
import matplotlib.pyplot as plt
from tensorboard.backend.event_processing import event_accumulator
import numpy as np
import matplotlib.ticker as ticker

def get_scalars(log_dir):

    ea = event_accumulator.EventAccumulator(log_dir,
        size_guidance={event_accumulator.SCALARS: 0})
    ea.Reload()

    scalars = {}
    for tag in ea.Tags()['scalars']:
        events = ea.Scalars(tag)
        steps = [e.step for e in events]
        values = [e.value for e in events]
        scalars[tag] = (steps, values)

    return scalars
```

```python
dirs = ['/data/save/lse/train_vdvae/train02.04-2',
        '/data/save/lse/train_vdvae/train02.04-3',
       ]

scalars = [get_scalars(dir) for dir in dirs]

plt.figure(figsize=[18, 5])
plt.plot(scalars[0]['distortion'][0], scalars[0]['distortion'][1])
plt.plot(scalars[1]['distortion'][0], scalars[1]['distortion'][1])
plt.grid()
plt.ylim([0.4, 0.7])
plt.show()
```

```python
plt.figure(figsize=[18, 5])
plt.plot(scalars[0]['rate'][0], scalars[0]['rate'][1])
plt.grid()
plt.show()
```

```python
plt.figure(figsize=[18, 5])
plt.plot(scalars[1]['rate'][0], scalars[1]['rate'][1])
plt.grid()
plt.ylim([0, 0.01])
plt.show()
```
