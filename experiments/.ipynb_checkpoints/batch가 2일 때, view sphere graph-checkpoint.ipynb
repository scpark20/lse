```python
import numpy as np
import os
import matplotlib.pyplot as plt

post_sigmas = np.arange(0.0, 5.0, 0.01)

def get_results(dir, dist):
    files = [os.path.join(dir, file) for file in os.listdir(dir) if file.startswith(dist)]
    print(len(files))
    # (n_files, post_sigma, k)
    results = np.array([np.load(file)['arr_0'] for file in files])
    # (post_sigma, k)
    results = np.mean(results, axis=0)
    # (k)
    results = post_sigmas[np.argmin(results, axis=0)]
    return results
```

```python
results = get_results('exps_min', 'sphere')
```

```python
import numpy as np
from scipy.special import gamma

def get_centroid(n):
    return gamma(0.5*n+1) / (np.sqrt(np.pi)*gamma(n/2+3/2))

ks = np.array([i for i in range(1, 64)])
centroid = get_centroid(ks)

centroid
```

```python
plt.figure(figsize=[18, 4])
plt.plot(results)
plt.plot(centroid)
plt.grid()
plt.show()
```
