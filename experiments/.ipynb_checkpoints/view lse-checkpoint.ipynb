```python
import numpy as np
import os
import matplotlib.pyplot as plt

ks = [1, 2, 4, 8, 16]
post_sigmas = np.arange(0.0, 5.0, 0.01)

def get_results(dir, dist):
    files = [os.path.join(dir, file) for file in os.listdir(dir) if file.startswith(dist)]
    print(len(files))
    # (n_files, post_sigma, N, k)
    results = np.array([np.load(file)['arr_0'] for file in files])
    # (post_sigma, N, k)
    results = np.mean(results, axis=0)
    # (N, k)
    results = post_sigmas[np.argmin(results, axis=0)]
    return results
```

```python
normal_results = get_results('exps_lse', 'normal')
uniform_results = get_results('exps_lse', 'uniform')
sphere_results = get_results('exps_lse', 'sphere')

# Improved plot design
plt.figure(figsize=[18, 5])

# Normal Results
plt.subplot(1, 3, 1)
plt.title('Normal', fontsize=14)
for i in range(normal_results.shape[1]):
    plt.plot(normal_results[:, i], label=str(ks[i]))
plt.grid(True, linestyle='--', alpha=0.5)
#plt.ylim([0, 3.0])
plt.legend(fontsize=10, loc='upper left')
plt.xticks([i for i in range(len(normal_results)) if i % 2 == 0])

# Uniform Results
plt.subplot(1, 3, 2)
plt.title('Uniform', fontsize=14)
for i in range(uniform_results.shape[1]):
    plt.plot(uniform_results[:, i], label=str(ks[i]))
plt.grid(True, linestyle='--', alpha=0.5)
#plt.ylim([0, 3.0])
plt.legend(fontsize=10, loc='upper left')
plt.xticks([i for i in range(len(uniform_results)) if i % 2 == 0])

# Sphere Results
plt.subplot(1, 3, 3)
plt.title('Sphere', fontsize=14)
for i in range(sphere_results.shape[1]):
    plt.plot(sphere_results[:, i], label=str(ks[i]))
plt.grid(True, linestyle='--', alpha=0.5)
#plt.ylim([0, 3.0])
plt.legend(fontsize=10, loc='upper left')
plt.xticks([i for i in range(len(sphere_results)) if i % 2 == 0])

# Show the improved plots
plt.tight_layout()
plt.show()
```
