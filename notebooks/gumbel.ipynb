```python
import numpy as np
import torch
```

```python
N = 10
eps = 1e-8
p = 1e-4
logit = np.log(p) - np.log(1-p)
logit
```

```python
u = torch.rand(N)
g = -torch.log(-torch.log(u+eps)+eps)
samples = torch.sigmoid((logit + g) / 1e-4)

import matplotlib.pyplot as plt
plt.figure(figsize=[18, 4])
plt.plot(samples.numpy())
plt.show()
```
