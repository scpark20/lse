```python
import numpy as np
```

```python
x = np.linspace(0, 5, 1000)

def func(x):
    return x ** 2

y = func(x)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x, y)
plt.grid()
plt.show()
```

```python
def func(x, tau=2):
    return np.abs(x) ** tau 

y = func(x, 0.1)

import matplotlib.pyplot as plt
plt.scatter(x, y)
plt.grid()
plt.ylim([0, 3])
plt.show()
```

```python
def func(x, tau=2):
    return np.abs(x) ** tau 

y = func(x, 10)

import matplotlib.pyplot as plt
plt.scatter(x, y)
plt.grid()
plt.ylim([0, 3])
plt.show()
```

```python
tau = 1e-1
y = np.where(x<1, x, func(x, 2/tau))
y = np.where(y>1, y, func(y, 2*tau))

import matplotlib.pyplot as plt
plt.scatter(x, y)
plt.grid()
plt.ylim([0, 3])
plt.show()
```
